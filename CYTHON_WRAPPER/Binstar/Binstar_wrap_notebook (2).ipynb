# explaining Binstar's second constructor:

### some important points:

1. The arguments in __cinit__ should be c++; otherwise, the constructor infront of **"NEW"** function will be a python object which can be wrong because new returns pointers.





2. There is an exception of a python object (_io) infront of __Cinit__ which was set because of using pythonic *"None"* in order to use the  **if conditions** for two constructors of the class.






3. The destructor is non-virtual that's why we recieve the warning, and it can be only resolved by changing the source code.



4. Because &params could'nt be set to zero, NULL, and None, we had to keep params as a non default arguments. However, the value of _io is responsible for changing between constructors.
